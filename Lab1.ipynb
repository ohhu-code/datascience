## LAB1

Write a function that takes the temperature Celsius as a list and outputs a list converted to Fahrenheit Write a function to convert a Fahrenheit temperature list to a Celsius temperature list.

오늘 하는 것은 제출할 필요는 없다. 불시에 오늘 한 것은 제출하세요 라고 할 수 있다. 그럴 때 제출.

IDS-02-2_Data_Analysis_Environments(Python_NumPy).pdf



```python
cel = [-20, -10, 0, 10, 20, 30, 40, 50]

def celsius_to_fahrenheit(celsius_list):
    return list((map)(lambda c : round(c * 9/5 + 32, 1), celsius_list))

def fahrenheit_to_celsius(fahrenheit_list):
    return list(map(lambda f: round((f - 32) * 5/9, 1), fahrenheit_list))

fahr = celsius_to_fahrenheit(cel)
print(fahr)

back_to_cel = fahrenheit_to_celsius(fahr)
print(back_to_cel)
```

## LAB2

Write a function that finds only palindrome in a given list and returns it as list.

```python
word = ['aaabbaaa', 'aabbbaaa', 'radar', '소주만병만주소']

def find_palindrome(word_list):
    return list(filter(lambda w: w == w[::-1], word_list))

result = find_palindrome(word)
print(result)
```

## LAB3
Write a function that extracts only vowels from a given word and returns it

모음만 추출해서 반환하라.  yield 를 써야한다.

```python
def extract_vowels(word):
  vowels = 'aeiouAEIOU'
  for char in word:
    if char in vowels:
      yield char

words = ['computer', 'python', 'sbs', 'sookmyung']

for word in words:
  result = list(extract_vowels(word))
  print(result)
```
